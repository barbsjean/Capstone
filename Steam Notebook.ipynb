```python
all_steam_data_MERGED = pd.read_json(r"C:\Users\Barbie\Desktop\DA11\Capstone\STEAM\API_CALLS_STEAM\COMBINED_TESTING_2.json")

all_steam_data_MERGED

# this is 73 seperate json files from 73 API Steam calls that I merged into one large json file using Notepad ++ 
```

```python
all_steam_data = all_steam_data_MERGED.transpose()  
all_steam_data     # using this to turn the columns into rows and the rows into columns

all_steam_data = all_steam_data.rename(
    columns = {'appid' : 'App_Id',
               'name' : 'Name',
               'developer' : 'Developer',
               'publisher' : 'Publisher',
               'score_rank' : 'Score_Rank',
               'positive' : 'Postive_Reviews',
               'negative' : 'Negative_Reviews',
               'userscore' : 'Userscore',
               'owners' : 'Owners',
               'average_forever' : 'Avg_Playtime_Lifetime',
               'average_2weeks' : 'Avg_Playtime_Two_Weeks',
               'median_forever' : 'Median_Playtime_Lifetime',
               'median_2weeks' : 'Median_Playtime_Two_Weeks',
               'price' : 'Current_Price',
               'initialprice' : 'Initial_Price',
               'discount' : 'Discount',
               'ccu' : 'Peak_CCU'})
all_steam_data     # renaming all the columns to make it flow better

all_steam_data = all_steam_data.reset_index()
all_steam_data     # resetting the index since it seemed to have its own index column identical to the "App_id" column

all_steam_data = all_steam_data.drop(columns = ['index'])
all_steam_data     # dropped the other index column to have the tradiation index column to run alongside the "App_id" column
```
